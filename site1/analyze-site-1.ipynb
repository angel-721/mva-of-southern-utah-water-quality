```python
### For site 1
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sn
```

```python
water_quality_df = pd.read_csv("../data/water-quality-site1.csv")
water_quality_columns_to_drop = ['Latitude', 'Longitude', 'Time', 'Date']

water_quality_df = water_quality_df.drop(columns=water_quality_columns_to_drop)
```

```python
### CLEAN DATA BY FILLING EMPTY COLUMS WITH MEAN VALUE

def fill_in_empty_cells_with_average(df):
    for col in df:    
        df[col].fillna(value=df[col].mean(), inplace=True)

fill_in_empty_cells_with_average(water_quality_df)
```

```python
### Make regression charts
```

```python
### Make Correlation Matrix
#water_quality_df = water_quality_df.drop(columns=['Time', 'Date'])
corr_matrix = water_quality_df.corr()

plt.subplots(figsize=(20,15))

figure = sn.heatmap(corr_matrix, annot=True).get_figure()
plt.show()
plt.savefig('water-site1.png')
```
